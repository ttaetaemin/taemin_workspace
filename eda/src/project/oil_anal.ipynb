```python
import pandas as pd
import matplotlib.pyplot as plt

#    (   )
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'

# CSV 
df = pd.read_csv('../data/2003~2025_oil.csv', encoding='euc-kr')

# ""  
df.set_index('', inplace=True)

# x    (    )
xtick_interval = len(df) // 20  # 20   

# ----------------------------
#     
# ----------------------------
plt.figure(figsize=(16, 6))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

plt.title('  ', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel(' ()', fontsize=12)
plt.xticks(ticks=range(0, len(df.index), xtick_interval), labels=df.index[::xtick_interval], rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------------
#     
# ----------------------------
for column in df.columns:
    plt.figure(figsize=(14, 5))
    plt.plot(df.index, df[column], linestyle='-', color='tab:blue', label=column)

    plt.title(f'{column}  ', fontsize=15)
    plt.xlabel('', fontsize=12)
    plt.ylabel(' ()', fontsize=12)
    plt.xticks(ticks=range(0, len(df.index), xtick_interval), labels=df.index[::xtick_interval], rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
