```python
import pandas as pd

pop_seoul = pd.read_excel("../data/01. Seoul_Population.xlsx", header=2, usecols="B,D,G,J,N")
pop_seoul.rename(columns={pop_seoul.columns[0]:"구별",
                          pop_seoul.columns[1]:"인구수",
                          pop_seoul.columns[2]:"한국인",
                          pop_seoul.columns[3]:"외국인",
                          pop_seoul.columns[4]:"고령자"},
                          inplace=True)
pop_seoul.head()

cctv_seoul = pd.read_csv("../data/01. Seoul_CCTV.csv", index_col=0)
cctv_seoul.reset_index(drop=True, inplace=True)
cctv_seoul.drop([0], inplace=True)
cctv_seoul.reset_index(drop=True, inplace=True)
cctv_seoul.head()

cctv_seoul.rename(columns={cctv_seoul.columns[0]:"구별"}, inplace=True)
cctv_seoul.rename(columns={cctv_seoul.columns[2]:"이전"}, inplace=True)
cctv_seoul.head()
# pop_seoul.drop([0], inplace=True)
```

```python

pop_seoul['구별'].unique()
pop_seoul['외국인비율'] = pop_seoul['외국인'] / pop_seoul['인구수'] * 100
pop_seoul['고령자비율'] = pop_seoul['고령자'] / pop_seoul['인구수'] * 100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by="고령자", ascending=False).head()
```

```python
left = pd.DataFrame({
    "key" : ["k0", "k4", "k2", "k3"],
    "A" : ["a0", "a1", "a2", "a3"],
    "B" : ["b0", "b1", "b2", "b3"]
})

right = pd.DataFrame({
    "key" : ["k0", "k1", "k2", "k3"],
    "C" : ["c0", "c1", "c2", "c3"],
    "D" : ["d0", "d1", "d2", "d3"]
})

left
```

```python
data_result = pd.merge(cctv_seoul, pop_seoul, on='구별')
del data_result['이전']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
del data_result['2019년']
del data_result['2020년']
del data_result['2021년']
del data_result['2022년']
del data_result['2023년']
data_result.set_index('구별', inplace=True)

data_result.to_csv('df.csv')
data_result.head()


```
