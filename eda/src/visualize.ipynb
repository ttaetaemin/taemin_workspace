```python
t = np.arange(0,12,0.01)
y = np.sin(t)

def drawgraph():
    plt.figure(figsize=(10,6))
    plt.plot(t, np.sin(t), label='sin')
    plt.plot(t, np.cos(t), label='cos')
    plt.grid()
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Example of sinwave')
    plt.show()

drawgraph()
```

```python
num = [1,7,9,16,36,39,45,45,46,48,51,100,101]
tem_y = [1] * len(num)
mean = np.mean(num)
med = np.median(num)
p1 = np.percentile(num, 25) 
p3 = np.percentile(num, 75)
```

```python
iqr = p3 - p1
upper = p3 + (1.5 * iqr)
lower = p1 - (1.5 * iqr)
```

```python
plt.figure(figsize=(12,5))
plt.scatter(num, tem_y)
plt.axvline(p1, color='black')
plt.axvline(med, color='red')
plt.axvline(p3, color='black')
plt.axvline(upper, color='black', ls='dashed')
plt.axvline(lower, color='black', ls='dashed')
plt.grid()
plt.show() 
```
